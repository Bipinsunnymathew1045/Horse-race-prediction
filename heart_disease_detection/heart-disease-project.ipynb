```python
y = data['Heart_Disease']
X = data.drop('Heart_Disease',axis = 1)
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le_X = {}
for i in X.select_dtypes(['object','bool']):
    le = LabelEncoder()
    X[i] = le.fit_transform(X[i])
    le_X[i]=le
le_y = LabelEncoder()
y = le_y.fit_transform(y)
```

```python
for col in le_X:
    print(col, le_X[col].classes_)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X[X.columns] = sc.fit_transform(X[X.columns])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
score = rfc.score(X_test,y_test)*100
print(f'Score : {score:.2f}%')
```

```python
from sklearn.metrics import accuracy_score
y_pred = rfc.predict(X_test)
ac_score = accuracy_score(y_pred,y_test)
print(f'accuracy_score :{ac_score*100:.2f}%')
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X_train,y_train)
```

```python
score = lor.score(X_test,y_test)*100
print(f'Score : {score:.2f}%')
```

```python
from sklearn.metrics import accuracy_score
y_pred = lor.predict(X_test)
ac_score = accuracy_score(y_pred,y_test)
print(f'accuracy_score :{ac_score*100:.2f}%')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
score = knn.score(X_test,y_test)*100
print(f'Score : {score:.2f}%')
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
acc_score= accuracy_score(y_test,y_pred)
print(f'accuracy_score : {acc_score*100:.2f}%')
```

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train,y_train)
```

```python
score = sv.score(X_test,y_test)*100
print(f'Score : {score:.2f}%')
```

```python
from sklearn.metrics import accuracy_score
y_pred = sv.predict(X_test)
acc_score= accuracy_score(y_test,y_pred)
print(f'accuracy_score : {acc_score*100:.2f}%')
```

```python
def my_predict(Age_in, Gender_in, Weight_in, Height_in, BMI_in, Smoking_in, Alcohol_Intake_in,
       Physical_Activity_in, Diet_in, Stress_Level_in, Hypertension_in, Diabetes_in,
       Hyperlipidemia_in, Family_History_in, Previous_Heart_Attack_in,
       Systolic_BP_in, Diastolic_BP_in, Heart_Rate_in, Blood_Sugar_Fasting_in,
       Cholesterol_Total_in):
    df1 = pd.DataFrame([{'Age': Age_in,
                         'Gender': Gender_in, 
                         'Weight': Weight_in,
                         'Height': Height_in, 
                         'BMI'   : BMI_in, 
                         'Smoking': Smoking_in, 
                         'Alcohol_Intake' : Alcohol_Intake_in,
                         'Physical_Activity' :  Physical_Activity_in, 
                         'Diet' : Diet_in,
                         'Stress_Level': Stress_Level_in, 
                         'Hypertension' : Hypertension_in,
                         'Diabetes' : Diabetes_in,
                         'Hyperlipidemia': Hyperlipidemia_in,
                         'Family_History' : Family_History_in,
                         'Previous_Heart_Attack': Previous_Heart_Attack_in,
                         'Systolic_BP' : Systolic_BP_in, 
                         'Diastolic_BP' : Diastolic_BP_in,
                         'Heart_Rate': Heart_Rate_in, 
                         'Blood_Sugar_Fasting': Blood_Sugar_Fasting_in,
                         'Cholesterol_Total' : Cholesterol_Total_in
                        }],index = [0])

    
    for i in df1.select_dtypes('object'):
      df1[i] = le_X[i].transform(df1[i])

    df1[X.columns] = sc.transform(df1[X.columns])

    return le_y.inverse_transform(lor.predict(df1))
```
